```python
import h5py
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

f = h5py.File('2017_2019_images_pv_processed.hdf5', 'r')
dset = f['trainval']['images_log']
total_count = dset.shape[0]

print(f"Ï¥ù Ïù¥ÎØ∏ÏßÄ Í∞úÏàò: {total_count}")

# 2. Ïù¥ÎØ∏ÏßÄÎ•º Î≥¥Ïó¨Ï£ºÎäî Ìï®Ïàò
def show_image(idx):
    img = dset[idx]

    # ÌòπÏãú ÏÉâÏù¥ Ïù¥ÏÉÅÌïòÎ©¥ ÏïÑÎûò Ï£ºÏÑù Ìï¥Ï†ú (BGR -> RGB)
    import cv2
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(3,3))
    plt.imshow(img.astype('uint8')) # Ï†ïÏàòÌòïÏúºÎ°ú Î≥ÄÌôò
    plt.axis('off')
    plt.title(f"Index: {idx}")
    plt.show()

# 3. Ïä¨ÎùºÏù¥Îçî ÎßåÎì§Í∏∞ (0Î≤àÎ∂ÄÌÑ∞ ÎßàÏßÄÎßâÍπåÏßÄ) 349372Ïû•
interact(show_image, idx=IntSlider(min=0, max=10000, step=1, value=0));
```
